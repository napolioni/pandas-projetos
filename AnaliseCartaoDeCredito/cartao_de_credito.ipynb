```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
data = pd.read_csv('MiBolsillo.csv', encoding="ISO-8859-1", sep=';')
```

```python
data.head()
```

```python
# mudança do nome das colunas para inglês
data.columns = [
    'id','branch_number','city','state','age','gender','total_credit_card_limit','current_available_limit' ,'date',
    'amount','category_expense','purchase_city','purchase_country']
```

```python
data
```

```python
data.date = pd.to_datetime(data.date,dayfirst=True)
data.head()
```

```python
customers = pd.DataFrame(data.id.unique())
customers.columns = ['old_customer_name']
customers['new_customer_name'] = np.arange(1, len(customers)+1)
customers
```

```python
# Alteração do id para entender os clientes com mais facilidade
data['id'].replace(customers['old_customer_name'].values, customers['new_customer_name'].values, inplace=True)
```

```python
data.info()
```

```python
for i in range(0,len(data.amount)):
    data.amount[i] = data.amount[i].replace('.','').replace(',','.')

#I export values containing '-' in the amount variable as df_amount_nan
df_amount_nan = data[data.amount == ' -   ']

# There are '-' values in the amount column. I am deleting theese rows because we can't convert it to the float type with these values
df = data[data.amount != ' -   ']

# Now the amount variable is ready to convert to float type.
data.amount = data.amount.astype(float)
```

```python
data.date = pd.to_datetime(data.date, dayfirst=True)
```

```python
dms = pd.get_dummies(data['gender'])
data = pd.concat([data, dms], axis=1)
data.drop(['gender', 'M'], axis=1, inplace=True)
data.rename(columns={'F': 'Famale'}, inplace=True)
data
```

```python
# Despesa total por categoria
categ = data.category_expense.value_counts().sort_values(ascending=False)


plt.figure(figsize=(15,10))
sns.barplot(x=categ.index, y=categ.values)
plt.xlabel('Categories')
plt.ylabel('Count')
plt.title('Categories Count')
plt.xticks(rotation=45);
```

```python
data.head()
```

```python
# Número de transações que cada cliente fez com cartão de crédito

freq = data.groupby('id')[['amount']].count().sort_values('amount', ascending=False)
freq.rename(columns={'amount': 'Frequency'}, inplace=True)
freq
```

```python
data['amount'].dtype
```

```python
# Gasto total de cada cliente

total = data.groupby('id')[['amount']].sum().sort_values('amount',ascending=False)
total.rename(columns={'amount': 'Total_spending'}, inplace=True)
total
```

```python
# Gasto total de cada cliente

plt.figure(figsize=(15,10))
sns.barplot(x=total.index, y=total['Total_spending'])
plt.xlabel('Customers')
plt.ylabel('Total Spending')
plt.title('Total Spending vs Customers')
plt.xticks(rotation=45)
```

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
data = pd.read_csv('MiBolsillo.csv', encoding="ISO-8859-1", sep=';')
```

```python
data.head()
```

```python
# mudança do nome das colunas para inglês
data.columns = [
    'id','branch_number','city','state','age','gender','total_credit_card_limit','current_available_limit' ,'date',
    'amount','category_expense','purchase_city','purchase_country']
```

```python
data
```

```python
data.date = pd.to_datetime(data.date,dayfirst=True)
data.head()
```

```python
customers = pd.DataFrame(data.id.unique())
customers.columns = ['old_customer_name']
customers['new_customer_name'] = np.arange(1, len(customers)+1)
customers
```

```python
# Alteração do id para entender os clientes com mais facilidade
data['id'].replace(customers['old_customer_name'].values, customers['new_customer_name'].values, inplace=True)
```

```python
data.info()
```

```python
data['amount'] = data['amount'].str.replace('\.', '').str.replace(',', '.')

data_amount_nan = data[data.amount == ' -   ']

df = data[data.amount != ' -   ']

data.amount = data.amount.astype(float)
```

```python
data.date = pd.to_datetime(data.date, dayfirst=True)
```

```python
dms = pd.get_dummies(data['gender'])
data = pd.concat([data, dms], axis=1)
data.drop(['gender', 'M'], axis=1, inplace=True)
data.rename(columns={'F': 'Famale'}, inplace=True)
data
```

```python
# Despesa total por categoria
categ = data.category_expense.value_counts().sort_values(ascending=False)


plt.figure(figsize=(15,10))
sns.barplot(x=categ.index, y=categ.values)
plt.xlabel('Categories')
plt.ylabel('Count')
plt.title('Categories Count')
plt.xticks(rotation=45);
```

```python
data.head()
```

```python
# Número de transações que cada cliente fez com cartão de crédito

freq = data.groupby('id')[['amount']].count().sort_values('amount', ascending=False)
freq.rename(columns={'amount': 'Frequency'}, inplace=True)
freq
```

```python
data['amount'].dtype
```

```python
# Gasto total de cada cliente

total = data.groupby('id')[['amount']].sum().sort_values('amount',ascending=False)
total.rename(columns={'amount': 'Total_spending'}, inplace=True)
total
```

```python
# Gasto total de cada cliente

plt.figure(figsize=(15,10))
sns.barplot(x=total.index, y=total['Total_spending'])
plt.xlabel('Customers')
plt.ylabel('Total Spending')
plt.title('Total Spending vs Customers')
plt.xticks(rotation=45)
```

```python
data['carnival'] = '0'
data["carnival"].replace(data[(data.date == '2020-02-24') | (data.date == '2020-02-25') | (data.date == '2019-03-04') | (data.date == '2019-03-05')]['carnival'],'1', inplace=True)


data['good_friday'] = '0'
data["good_friday"].replace(data[(data.date == '2019-04-19') | (data.date == '2020-04-10')]['good_friday'],'1', inplace=True)

data['christmas'] = '0'
data["christmas"].replace(data[(data.date == '2019-12-25') | (data.date == '2020-12-25')]['christmas'],'1', inplace=True)

data['corpus_christi'] = '0' 
data["corpus_christi"].replace(data[(data.date == '2019-06-20') | (data.date == '2020-06-11')]['corpus_christi'],'1', inplace=True)

data['new_year'] = '0'
data["new_year"].replace(data[(data.date == '2019-01-01') | (data.date == '2020-01-01')]['new_year'],'1', inplace=True)


data['black_friday'] = '0'
data["black_friday"].replace(data[(data.date == '2019-11-29') | (data.date == '2020-11-27')]['black_friday'],'1', inplace=True)

data['halloween'] = '0'
data["halloween"].replace(data[(data.date == '2019-10-31') | (data.date == '2020-10-31')]['halloween'],'1', inplace=True)
```

```python
# visualização do total de transações nos dias especiais

special_days = ['carnival', 'good_fryday', 'christmas','corpus_christi', 'new_year', 'black_friday', 'halloween' ]

counts = []

for i in special_days:
    counts.append(data[i].values_counts()[1])
special_days_dict = dict(zip(special_days, counts))


plt.figure(figsize=(15,10))
sns.barplot(x=list(special_days_dict.keays()), y=list(special_days_dict.values()))
plt.xlabel('Special Days')
plt.ylabel('Toatal Transation')
plt.title('Total Transation of Special Days')
plt.xticks(rotation=45);
```

```python
# Frequência de uso

covid_freq = covid.groupby('id')[['age']].count().sort_values('age', ascending=False)
covid_freq.columns = ['frequency']
print('average number os transactions frequency: ', round(covid_freq.frequency.mean(), 2))
print('max number os transactions frequency: ', covid_freq.frequency.max())
print('min number os transactions frequency: ', covid_freq.frequency.min())
```

```python
plt.figure(figsize=(15,10))
sns.barplot(x=covid_freq.index, y=covid_freq['frequency'])
plt.xlabel('Customers')
plt.ylabel("frequency of use")
plt.title('The frequency of transactions made by customers during the covid')
plt.xticks(rotation=45);
```
